```python
from fuzzywuzzy import fuzz, process
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = 'data/Assets/Brewers_Association_Data_2015-1-1-1/2015_craft_volume.csv'

df = pd.read_csv(filepath, header=1)
breweries = pd.read_csv('data/master_untappd_brewery_data.csv')
```

```python
def mop_the_brewhouse(brewery_name):
    cleaned = brewery_name.replace('\n','').replace('\t', '')
    return cleaned
```

```python
breweries['brewery_name_cleaned'] = breweries['brewery_name'].apply(mop_the_brewhouse)
```

```python
breweries['brewery_name_cleaned'][0:5]
```

```python
breweries
```

```python
brewery_id_dict = {}
    
for index, row in breweries.iterrows():
    brewery_id_dict[row['brewery_name_cleaned']] = row['brewery_id']
```

```python
brewery_id_dict
```

```python
choices = brewery_id_dict.keys()
process.extractOne('Magic Hat Brewing Company', choices)[0]
```

```python
def get_id(craft_brewery):
    match = process.extractOne(craft_brewery, choices)
    if match[1] > 80:
        return brewery_id_dict[match[0]]
    else:
        print(match, 'not in list')
```

```python
df['brewery_id'] = df['Craft Brewer Name'].apply(get_id)
```
